# CONVERT THE TOTAL MASK FOLDER TO BINARY CLASSIFICATION

```python
import os
import numpy as np
from PIL import Image

# Define paths to input image directory and output mask directory
input_dir = r"D:\dataset 2\new data\New folder\images"  # Path to your images dataset
output_dir =r"D:\dataset 2\new data\New folder\masks"  # Path to save the binary masks

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Define the color that represents deforested areas (e.g., red)
deforested_color = np.array([255, 0, 0])  # Red
forest_color = np.array([0, 255, 0])  # Green (optional if you want to map forest as well)

# Loop through all image files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.png') or filename.endswith('.jpg'):
        # Load the image
        image_path = os.path.join(input_dir, filename)
        image = Image.open(image_path)
        image_array = np.array(image)

        # Initialize the binary mask with zeros (same height and width as the image)
        binary_mask = np.zeros((image_array.shape[0], image_array.shape[1]), dtype=np.uint8)

        # Set deforested areas to 1 based on color
        # Use np.all along the last axis to compare RGB values
        binary_mask[np.all(image_array == deforested_color, axis=-1)] = 1

        # Optionally, set forest areas to 0 (not necessary since the mask is initialized with zeros)
        # binary_mask[np.all(image_array == forest_color, axis=-1)] = 0  # This line can be skipped

        # Convert the binary mask to a PIL image
        binary_mask_image = Image.fromarray(binary_mask * 255)  # Multiply by 255 to get a proper grayscale image

        # Save the binary mask image
        binary_mask_path = os.path.join(output_dir, filename)
        binary_mask_image.save(binary_mask_path)

        print(f"Saved binary mask for {filename} at {binary_mask_path}")
```

```python
import matplotlib.pyplot as plt
# Show the binary mask
plt.imshow(binary_mask_image, cmap='Gray')
plt.title('Binary Mask')
plt.show()

# Print to check
print("Binary mask created with shape:",binary_mask_image)
```
